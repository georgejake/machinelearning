```python
df_homeprices = pd.read_csv("home_prices.csv")
print(df_homeprices.head())
# We are creating a model for home prices
model = LinearRegression()
# Preparing the features
X = df_homeprices[ ["area_sqr_ft"]]
y = df_homeprices["price_lakhs"]
model.fit(X,y)
print(f"The slope is {model.coef_} and intercept is {model.intercept_}")
```

#### For the same dataset we are creating our own model based on gardient descent

We start with an aribitary value of m(slope) and b (intercept). We calculate the predicted value and the MSE for that aribitary value of m & b. Continue with multiple epochs (iterations) till we get a minimum value of MSE. The learning rate $\alpha$ is set as .05
The next value of m and b is calculated using partial derivatives

```python
m = 0
b = 0
learning_rate = .05
epoch = 1000
# Due to the magnitude difference between area v/s price ,we have to scale the values
# For scaling the method used is 
# (X - X(min))/( X(max)- X(min) )
X = np.array(df_homeprices["area_sqr_ft"])
Y = np.array(df_homeprices["price_lakhs"])
x_min,x_max = X.min(),X.max()
y_min,y_max = Y.min(),Y.max()

x_scaled = (X-x_min)/(x_max-x_min)
y_scaled = (Y-y_min)/(y_max-y_min)
# print(x_scaled,y_scaled)

print(y_predicted)
for i in range(epoch):
    # y_predicted is also a numpy array [all are array operations using numpy
    y_predicted = m*x_scaled + b
    error = y_scaled - y_predicted
    mse = np.mean(error ** 2)
    # new value of m and c (using the partial derivatives dm=change of MSE w.r.t m keeping b constant ,
    # db = change of MSE w.r.t b keeping m constant )
    db = -2*np.mean(error)
    dm = -2*np.mean(x_scaled*error)
    m -= learning_rate*dm
    b -= learning_rate*db

# Scale back the coefficients to original scale
b_original = b * (y_max - y_min) + y_min - m * (y_max - y_min) * x_min / (x_max - x_min)
m_original = m * (y_max - y_min) / (x_max - x_min)
print(f"The slope is {m_original} and intercept is {b_original}")
```
